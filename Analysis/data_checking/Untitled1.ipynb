```python
import os
import pydicom



import nibabel as nib
import numpy as np

import matplotlib.pyplot as plt
```

```python
path = '/mnt/neurocube/local/serenceslab/maggie/shapeDim/DataPreproc/S02/'

fn = '01_REG_MC_DET_001.nii.gz'
filename = os.path.join(path, fn)
```

```python
im = nib.load(filename)
im = np.array(im.get_fdata())

im.shape
```

```python
im.shape
```

```python
plt.figure(figsize=(12,12))
t2plot = [0, 50, 100, 150]

for ti, tt in enumerate(t2plot):

    plt.subplot(2,2,ti+1)
    # xx = 50;
    # yy = 50;
    zz = 38;


    slic = im[:,:,zz, tt]
    slic = np.flipud(slic.T)
    plt.imshow(slic, cmap='gray')
    plt.xticks([])
    plt.yticks([])
    
    plt.title('t=%d'%(tt))

folder2save = '/mnt/neurocube/local/serenceslab/maggie/shapeDim/Analysis/data_checking/'
fn2save = os.path.join(folder2save, 'examples.pdf')
print(fn2save)
plt.savefig(fn2save)
```
