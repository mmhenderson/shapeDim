```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys
import pandas as pd

root = '/usr/local/serenceslab/maggie/shapeDim/'

sys.path.append(os.path.join(root, 'Analysis'))
from code_utils import file_utils
```

```python
sublist = np.arange(1,8)
idx = -1
sdat = pd.DataFrame(columns = ['subject','run','num_events','hit_rate','num_FA'])

for ss in sublist:

    rc = 0
    for se in [1,2,3,4]:

        bdir = os.path.join(root, 'DataBehavior', 'S%02d'%ss, 'Session%d'%se)
        if not os.path.exists(bdir):
            continue
        silh_files = os.listdir(bdir)
        silh_files = [ff for ff in silh_files if 'Silh' in ff]

        if len(silh_files)>0:
            assert(len(silh_files)==1)
            TheData = file_utils.load_mat_behav_data(os.path.join(bdir, silh_files[0]))
            print('found %d runs for S%02d'%(len(TheData), ss))
            for rr in range(len(TheData)):
                rc+=1; idx+=1
                dat = TheData[rr]['data']

                sdat = sdat.append(pd.DataFrame({'subject': ss, \
                                         'run': rc, \
                                         'num_events': dat['EvtCount'], \
                                         'hit_rate': dat['hit_rate'], \
                                         'num_FA': dat['FACount']}, index=[idx]))
```

#### Print some simple statistics

```python
subj_avg_hitrates = [np.mean(sdat['hit_rate'][sdat['subject']==ss]) for ss in sublist]
m = np.mean(subj_avg_hitrates)
s = np.std(subj_avg_hitrates)
print('average hit rate (mean +/- std):')
print('%.2f +/- %.2f'%(m, s))
```

```python
subj_avg_fa = [np.mean(sdat['num_FA'][sdat['subject']==ss]) for ss in sublist]
m = np.mean(subj_avg_fa)
s = np.std(subj_avg_fa)
print('average num false alarms (mean +/- std):')
print('%.2f +/- %.2f'%(m, s))
```

```python
evtcount = sdat['num_events']
m = np.mean(evtcount)
mi = np.min(evtcount)
ma = np.max(evtcount)
print('average event count (mean, [min, max]):')
print('%d, [%d, %d]'%(m, mi, ma))
```
